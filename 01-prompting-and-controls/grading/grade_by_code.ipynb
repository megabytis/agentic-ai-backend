````python
import json
import ast
import re
````

````python
def validate_json(text):
    try:
        json.loads(text)
        return 10
    except Exception:
        return 0
````

````python
def validate_python(text):
    try:
        ast.parse(text)
        return 10
    except Exception:
        return 0
````

````python
def validate_regex(text):
    try:
        re.compile(text)
        return 10
    except Exception:
        return 0
````

````python
def grade_syntax(output_item, test_case):
    fmt = test_case.get("format", "").lower()

    output_text=output_item['output']
    clean_output=output_text

    if fmt == "json":
        return validate_json(clean_output)
    elif fmt == "python":
        return validate_python(clean_output)
    elif fmt in ["regex", "regexp"]:
        return validate_regex(clean_output)
    else:
        return 0
````

````python
def grade_syntx_all(outputs_list,test_cases_list):
    """ Grade all test cases """
    results=[]

    for i in range(len(test_cases_list)):
        test_case=test_cases_list[i]
        output_item=outputs_list[i]
        score=grade_syntax(output_item,test_case)
        results.append(score)
    return results
````

````python
def clean_code_output(text):
    """Remove markdown code block markers"""
    text = text.strip()
    if text.startswith('```') and text.endswith('```'):
        # Remove first and last lines (the ``` markers)
        lines = text.split('\n')
        if len(lines) > 2:
            return '\n'.join(lines[1:-1]).strip()
    return text
````

````python
with open("../eval/outputs_v1.json", "r") as f:
    output = json.load(f)

with open("../eval/tasks.json", "r") as f:
    test_case = json.load(f)

score = grade_syntx_all(output,test_case)
score
````
