```python
import json
import ast
import re
```

```python
def validate_json(text):
    try:
        json.loads(text)
        return 10
    except Exception:
        return 0
```

```python
def validate_python(text):
    try:
        ast.parse(text)
        return 10
    except Exception:
        return 0
```

```python
def validate_regex(text):
    try:
        re.compile(text)
        return 10
    except Exception:
        return 0
```

```python
def grade_syntax(output: str, test_case: dict) -> int:
    fmt = test_case.get("format", "").lower()

    if fmt == "json":
        return validate_json(output)
    elif fmt == "python":
        return validate_python(output)
    elif fmt in ["regex", "regexp"]:
        return validate_regex(output)
    else:
        return 0
```
