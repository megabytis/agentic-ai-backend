```python
import os
from dotenv import load_dotenv
load_dotenv()

from openai import OpenAI
from langchain.messages import HumanMessage,AIMessage

client = OpenAI(
    api_key=os.getenv("OPENROUTER_API_KEY"),
    base_url="https://openrouter.ai/api/v1"
)

completion = client.chat.completions.create(
    model="openai/gpt-oss-120b:free",
    messages=[
        {
            "role": "user",
            "content": "What is GenAI tell me in one line in simple english within 50 words ?",
        },
    ],
)

print(completion)
```

```python
from dotenv import load_dotenv
load_dotenv()
import os
from openai import OpenAI

client = OpenAI(
    api_key=os.getenv("OPENROUTER_API_KEY"),
    base_url="https://openrouter.ai/api/v1/"
)

response = client.responses.create(
    model="nvidia/nemotron-3-nano-30b-a3b:free",
    # instructions="You are a coding assistant that talks like a pirate.",
    input="give me a JSON list which contains 5 country's name, their capital, their currency, their president/primeminister name & their country's best dish name. the answer should strictly in JSON format, no extra talk, no BS, no fluff, straight answer .",
)

print(response.output_text)
```
